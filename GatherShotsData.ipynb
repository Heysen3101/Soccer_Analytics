```python
#lets filter passes 
#and focus on the location

shots = shots[['timestamp','player_name','possession_team_name','shot_outcome_name', 'location','period']]
shots.head(30)
```

```python
# Split the 'location' into 'x' and 'y'
shots['x'] = shots['location'].apply(lambda loc: loc[0])
shots['y'] = shots['location'].apply(lambda loc: loc[1])

# Drop the original 'location' and 'pass_end_location' columns if they are no longer needed
shots.drop(['location'], axis=1, inplace=True)

shots
```

```python
import pandas as pd

# Step 1: Convert 'timestamp' to datetime and extract minutes and seconds
shots['timestamp'] = pd.to_datetime(shots['timestamp']).dt.strftime('%M:%S')

# Step 2: Adjust the minutes based on the period
# This creates a function that adds the appropriate number of minutes
def adjust_time(row):
    if row['period'] == 2:
        added_time = 45
    elif row['period'] == 3:
        added_time = 90
    elif row['period'] == 4:
        added_time = 105
    else:
        added_time = 0

    # Extract minutes and seconds from the timestamp
    minutes, seconds = map(int, row['timestamp'].split(':'))
    # Add the appropriate amount of time
    total_minutes = minutes + added_time
    # Return the adjusted time in the format MM:SS
    return f'{total_minutes:02d}:{seconds:02d}'

# Apply the function to each row
shots['adjusted_timestamp'] = shots.apply(adjust_time, axis=1)
```
