```python
# load some important libraries
import pandas as pd
from pandas import json_normalize
import requests
```

```python
#load competition file

competitions = requests.get("https://raw.githubusercontent.com/statsbomb/open-data/master/data/competitions.json")
competitions = competitions.json()
```

```python
#print all the competiotion with competition id and season_id

for competition in competitions:
    print(competition['competition_name'], competition['season_name'], competition['competition_id'], competition['season_id'])
```

```python
#lets focus fifa world cup

competition_id = 43
season_id = 106
```

```python
#lets get all the matches
matches = requests.get('https://raw.githubusercontent.com/statsbomb/open-data/master/data/matches/'+str(competition_id)+'/'+str(season_id)+'.json')
matches = matches.json()
```

```python
matches
```

```python
#print all the games in this competitions with their scores and match ids

for match in matches:
    print(match['home_team']['home_team_name'], match['away_team']['away_team_name']+\
          " " + str(match['home_score']) + ":" + str(match['away_score']) + "(match_id is" + str(match['match_id']) + ")")
    
```

```python
#lets explore france

team = "France"
```

```python
#print all france matches with their score and match id

for match in matches:
    if match['home_team']['home_team_name'] == team or match['away_team']['away_team_name'] == team:
      print(match['home_team']['home_team_name'], match['away_team']['away_team_name']+\
          " " + str(match['home_score']) + ":" + str(match['away_score']) + "(match_id is" + str(match['match_id']) + ")")
    
```

```python
#lets focus on argentina passes

Argentina.head(5)
```

```python
#lets filter passes 
#and focus on the location

Argentina_passes = Argentina.loc[Argentina['type_name'] == 'Pass']
Argentina_passes = Argentina_passes[['player_name', 'pass_recipient_name', 'location', 'pass_end_location',]]
Argentina_passes.head(30)
```

```python
#creating another column
import pandas as pd
import numpy as np

# Assuming df is your DataFrame
# You can use `np.where` to conditionally fill the new column
Argentina_passes['pass_outcome'] = np.where(Argentina_passes['pass_recipient_name'].isna(), 'unsuccessful', 'successful')
```

```python
Argentina_passes.head(5)
```

```python
#only lionel messi passes
messi_passes = Argentina_passes[Argentina_passes['player_name'] == 'Lionel Andrés Messi Cuccittini']
messi_passes
```

```python
#changing lionel andres messi name to messi
messi_passes.loc[messi_passes['player_name'] == 'Lionel Andrés Messi Cuccittini', 'player_name'] = 'Lionel Messi'

```

```python
messi_passes
```

```python
#we no longer need pass_recipientname column
messi_passes.drop('pass_recipient_name', axis=1, inplace=True)
```

```python

messi_passes
```

```python
# Split the 'location' into 'x' and 'y'
messi_passes['x'] = messi_passes['location'].apply(lambda loc: loc[0])
messi_passes['y'] = messi_passes['location'].apply(lambda loc: loc[1])

# Split the 'pass_end_location' into 'end_x' and 'end_y'
messi_passes['end_x'] = messi_passes['pass_end_location'].apply(lambda loc: loc[0])
messi_passes['end_y'] = messi_passes['pass_end_location'].apply(lambda loc: loc[1])

# Drop the original 'location' and 'pass_end_location' columns if they are no longer needed
messi_passes.drop(['location', 'pass_end_location'], axis=1, inplace=True)

messi_passes
```
