```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
import seaborn as sns
from mplsoccer import VerticalPitch
import matplotlib as mpl
```

```python
#read the data
df = pd.read_csv('messi_passes.csv')
df
```

```python
# Set up the pitch
pitch = Pitch(pitch_type='statsbomb', pitch_color='grass', line_color='#c7d5cc', stripe=True)##22312b
fig, ax = pitch.draw(figsize=(10, 6), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#22312b')

#plt.gca().invert_yaxis()

# Create the heat map using the corrected syntax
kde = sns.kdeplot(
        x=df['x'],
        y=df['y'],
        shade= True,
        shade_lowest = False,
        alpha=.4,
        n_levels=8,
        cmap='magma'
    )

for x in range(len(df['x'])):
    if df['pass_outcome'][x] == 'successful':
        plt.plot((df['x'][x],df['end_x'][x]),(df['y'][x],df['end_y'][x]),color='yellow')
        plt.scatter(df['x'][x],df['y'][x],color='yellow')
    if df['pass_outcome'][x] == 'unsuccessful':
        plt.plot((df['x'][x],df['end_x'][x]),(df['y'][x],df['end_y'][x]),color='red')
        plt.scatter(df['x'][x],df['y'][x],color='red')

#plt.xlim(0,120)
#plt.ylim(0,80)
```

```python
# Set up the pitch
pitch = VerticalPitch(pitch_type='statsbomb', pitch_color='grass', line_color='#c7d5cc', stripe=True)##22312b
fig, ax = pitch.draw(figsize=(10, 6), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#22312b')

#plt.gca().invert_yaxis()

# Create the heat map using the corrected syntax
kde = sns.kdeplot(
        x=df['y'],
        y=df['x'],
        shade= True,
        shade_lowest = False,
        alpha=.4,
        n_levels=8,
        cmap='magma'
    )

for y in range(len(df['y'])):
    if df['pass_outcome'][y] == 'successful':
        plt.plot((df['y'][y],df['end_y'][y]),(df['x'][y],df['end_x'][y]),color='yellow')
        plt.scatter(df['y'][y],df['x'][y],color='yellow')
    if df['pass_outcome'][y] == 'unsuccessful':
        plt.plot((df['y'][y],df['end_y'][y]),(df['x'][y],df['end_x'][y]),color='red')
        plt.scatter(df['y'][y],df['x'][y],color='red')
plt.title('Messi Pass/Heatmap vs France', fontsize=12, color='white')
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='Yellow', edgecolor='none', label='Successful'),
    Patch(facecolor='red', edgecolor='none', label='UnSuccessful')
]
plt.legend(handles=legend_elements, loc='lower left', fontsize='xx-small', facecolor='white', title='Pass Color')
#plt.xlim(0,120)
#plt.ylim(0,80)
```
